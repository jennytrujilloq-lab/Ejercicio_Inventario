```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("notas_estudiantes.csv")

sns.barplot(x='asignatura', y='nota', data=df)
plt.title("Nota media por asignatura")
plt.ylim(0, 10)
plt.grid(True, axis='y', linestyle='--', alpha=0.5)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Leemos el archivo
df = pd.read_csv("notas_estudiantes.csv")

# --- IMPORTANTE ---
# Como tu CSV no tiene 'edad' ni 'asistencia', las creamos
# temporalmente para que el código de la diapositiva funcione:
np.random.seed(42)
df['edad'] = np.random.randint(15, 19, size=len(df))
df['asistencia'] = np.random.randint(70, 100, size=len(df))
# ------------------

# CÓDIGO EXACTO DE LA DIAPOSITIVA:
correlacion = df[['nota', 'edad', 'asistencia']].corr()

sns.heatmap(correlacion, annot=True, cmap='coolwarm')
plt.title("Correlación entre nota, edad y asistencia")
plt.show()
```
