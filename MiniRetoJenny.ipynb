Mini Reto con bucle de Inventarios


```python
inventario = int(input("Ingresa la cantidad inicial de pares de zapatos: "))
transacciones = 0
while inventario > 0:
    vendidos = int(input(f"¿Cuántos pares de zapatos se vendieron? (quedan {inventario}): "))
    if vendidos > inventario:
        print(f"No puedes vender más de {inventario} pares. Intenta de nuevo.")
        continue
    inventario -= vendidos
    transacciones += 1
    print(f"Inventario restante: {inventario}")

print(f"\n¡Inventario agotado!")
print(f"Total de transacciones realizadas: {transacciones}")
```
